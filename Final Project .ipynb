## Persamaan konduksi panas dalam sebuah tongkat

Untuk final project kami, kami diberikan topik untuk mencari pergantian suhu di sebuah tongkat dalam waktu yang
ditentukan. Contohnya, adalah jika sebuah tongkat yang berukuran 100cm mempunyai suhu sebesar 100°C akan didinginkan 
selama 10 detik. Tujuan dari project ini adalah untuk menemukan pergantian suhu setiap seperseribu detik selama 10 
detik ini. 
Jadi pertama, kami akan menggunakan numpy dan matplotlib untuk menghitung dan mencari hasil dari penelitian ini.
Ini dapat dilakukan dengan cara:

    import numpy as np;
    import matplotlib.pyplot as plt

Dengan menggunakan kata ‘import’ dari awal dokumen, kami secara langsung sudah 
diperbolehkan untuk menggunakan segala fungsi dari numpy dan matplotlib untuk menghitung berbagai rumus.
Selanjutnya, kami membuat sebuah fungsi untuk menghitung suhu pada setiap pergantian seperseribu detik, seperti berikut:

    def calculate(pointinmid,calcmatrix,time,coeff,currtime=1):

        for cal in range(1,pointinmid-1):
            calcmatrix[currtime][cal]=calcmatrix[currtime-1][cal] + coeff * (calcmatrix[currtime-1][cal-1] + 
            calcmatrix[currtime-1][cal+1] - (2 * calcmatrix[currtime-1][cal]))
            print(calcmatrix[currtime][cal])
        print("----------------------------------------------")
        currtime+=1
        if (currtime <= time):
            return calculate(pointinmid,calcmatrix,time,coeff,currtime)
        else:
            return calcmatrix

Dengan menaruh pointinmid,calcmatrix,time,coeff,currtime=1, sebagai parameter di dalam fungsinya , kita 
akan mengimplementasikan rumus

$$
Tl+1i= Tli+ λTli+1− 2Tli+ Tli−1
$$

dimana $$λ = k t/(x)2.$$


Selanjutnya kami membuat sebuah ‘user input interaction’ dimana pengguna dapat menentukan kondisi yang akan 
menentukan hasil dari penelitian ini, seperti berikut:

    coeff=float(input("Thermal conductivity :"))
    x = float(input("Length of the rod :"))
    fronttemp=float(input("Front temperature :"))
    backtemp=float(input("Back Temperature :"))
    time=int(input("Time :"))
    pointinmid=int(input("Total of Points between front and back:"))

Input diatas akan menentukan faktor2 dalam perbedaan suhu di sebuah tongkat. 
Saat menjalankan kode, setelah kami mengetikkan input berikut, titik intmid akan bertambah, deltax nilai x membagi 
pointintmid. Untuk coordx, kami menggunakan larik untuk menghitung delta kali batang dengan 
menggunakan loop dari (pointinmid + 1). Fronttemp adalah titik awal dalam sebuah tongkat dan juga 
suhu awal pada sebuah tongkat. 
Backtemp adalah titik akhir dalam sebuah tongkat dan juga suhu terakhir pada sebuah tongkat. 
Untuk mendapat hasil akhir, kami harus mengimplementasikan beberapa hal tambahan seperti berikut:

    pointinmid+=1
    deltax=x/pointinmid
    coordx=np.array([deltax*cordx for cordx in range(pointinmid+1)])
    pointinmid+=1
    initial=np.array([[0.0 for i in range(pointinmid)] for j in range(time+1)])
    for l in range (time+1):
        initial[l][0]=fronttemp
        initial[l][-1]=backtemp
    xtest=calculate(pointinmid,initial,time,coeff)
